```python
import numpy as np
import pandas as pd

import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
df = pd.read_csv('/content/ecommerce_newest_1.csv')
```

```python
df.info()
```

calculate the total revenue for all product per year

```python
total_revenue_per_year = df.groupby('year')['revenue'].sum()
total_revenue_per_year
```

```python
df['cost*num_of_item'] = df['cost'] * df['num_of_item']
```

```python
df['sale_price*num_of_item'] = df['sale_price'] * df['num_of_item']
```

calculate total cost for all product per year

```python
total_cost_per_year = df.groupby('year')['cost*num_of_item'].sum()
total_cost_per_year
```

calculate total profit margin per year

```python
total_profit_margin_per_year = ((total_revenue_per_year - total_cost_per_year) / total_revenue_per_year) * 100
total_profit_margin_per_year
```

```python
cost_increase_10 = total_cost_per_year * 1.1
new_profit_margin_10 = ((total_revenue_per_year - cost_increase_10) / total_revenue_per_year) * 100
new_profit_margin_10
```

```python
cost_decrease_10 = total_cost_per_year * 0.9
new_profit_margin_decrease_10 = ((total_revenue_per_year - cost_decrease_10) / total_revenue_per_year) * 100
new_profit_margin_decrease_10
```

```python
price_increase_10 = total_revenue_per_year * 1.1
new_profit_margin_price_increase_10 = ((price_increase_10 - total_cost_per_year) / price_increase_10) * 100
new_profit_margin_price_increase_10
```
