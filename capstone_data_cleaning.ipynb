```python
import numpy as np
import pandas as pd

import statsmodels.api as sm
from sklearn.linear_model import LinearRegression

from matplotlib import pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
df = pd.read_csv('/content/ecommerce_newest.csv')
df.head()
```

```python
df.info()
```

```python
df['created_at'] = pd.to_datetime(df['created_at'], format='mixed', utc=True)
df['created_at'].head()
```

```python
df['year'] = df['created_at'].dt.year
df.head()
```

```python
df['num_of_item'] = df['num_of_item'].convert_dtypes('int64')
df['order_id'] = df['order_id'].convert_dtypes('int64')
df['year'] = df['year'].convert_dtypes('int64')
df['product_id'] = df['product_id'].convert_dtypes('int64')
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(0, inplace=True)
```

```python
df.isnull().sum()
```

1. Identify the best and the worst-selling products for each year.

```python
product_sales_yearly = df.groupby(['product_id', 'product_name', 'year'])['num_of_item'].sum().reset_index()
product_sales_yearly.head()
```

```python
best_selling_products_yearly = product_sales_yearly.groupby('year')['num_of_item'].idxmax()
best_selling_products_yearly = product_sales_yearly.loc[best_selling_products_yearly]
best_selling_products_yearly
```

```python
worst_selling_products_yearly = product_sales_yearly.groupby('year')['num_of_item'].idxmin()
worst_selling_products_yearly = product_sales_yearly.loc[worst_selling_products_yearly]
worst_selling_products_yearly
```

2. Calculate revenue for each product.

```python
df['revenue'] = df['num_of_item'] * df['sale_price']
df.head()
```

3. Calculate the total revenue for each product and determine the most profitable and the least profitable product for each year.

```python
product_revenue_yearly = df.groupby(['product_id', 'product_name', 'year'])['revenue'].sum().reset_index()
product_revenue_yearly.head()
```

```python
most_profitable_products_yearly = product_revenue_yearly.groupby('year')['revenue'].idxmax()
most_profitable_products_yearly = product_revenue_yearly.loc[most_profitable_products_yearly]
most_profitable_products_yearly
```

```python
least_profitable_products_yearly = product_revenue_yearly.groupby('year')['revenue'].idxmin()
least_profitable_products_yearly = product_revenue_yearly.loc[least_profitable_products_yearly]
least_profitable_products_yearly
```

4. Analyze Gender and Age Preferences for Products

```python
gender_preferences = df.groupby(['product_id', 'product_name', 'gender'])['num_of_item'].sum().reset_index()
gender_preferences
```

create age groups

```python
bins = [0, 18, 25, 35, 45, 60, 100]
labels = ['0-18', '19-25', '26-35', '36-45', '46-60', '61+']
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels, right=False)
df.head()
```

```python
age_preferences = df.groupby(['product_id', 'product_name', 'age_group'], observed=True)['num_of_item'].sum().unstack()
```

```python
age_preferences.fillna(0, inplace=True)
age_preferences
```

bonus question: perform a what if analysis for profit margin for each product( assessing how different profit margins would impact the overall revenue and profitability of each product.

```python
df['profit_margin'] = ((df['revenue'] - (df['num_of_item'] * df['cost'])) / df['revenue']) * 100
df.head()
```

what if scenerios:
1. increase cost by 10%

```python
df['cost_increase_10'] = df['cost'] * 1.10
df['new_profit_margin_cost_increase_10'] = ((df['revenue'] - (df['num_of_item'] * df['cost_increase_10'])) / df['revenue']) * 100
df.head()
```

2. Decrease cost by 10%

```python
df['cost_decrease_10'] = df['cost'] * 0.90
df['new_profit_margin_cost_decrease_10'] = ((df['revenue'] - (df['num_of_item'] * df['cost_decrease_10'])) / df['revenue']) * 100
df.head()
```

3. increase selling price by 10%

```python
df['selling_price_increase_10'] = df['sale_price'] * 1.10
df['new_revenue_price_increase_10'] = df['num_of_item'] * df['selling_price_increase_10']
df['new_profit_margin_price_increase_10'] = ((df['new_revenue_price_increase_10'] - (df['num_of_item'] * df['cost'])) / df['new_revenue_price_increase_10']) * 100
df.head()
```
